Labeling CPU dengan range 1-9

```python
input_file = "/content/cleaned.csv"
output_file = "cleaned_scored_cpu.csv"

def processor_score(processor_name):
  if processor_name.startswith("AMD Ryzen"):
    if processor_name == "AMD Ryzen Z1 Extreme":
      return 5
    elif len(processor_name.split()) > 1:
      core_number_str = processor_name.split()[2]
      core_number = int(core_number_str)
    else:
      core_number = 2
    if core_number == 3:
      return 3
    elif core_number in (5, 7):
      return core_number
    else:
      return 2
  elif processor_name.startswith("Intel Core"):
    match = re.search(r"\d+", processor_name)
    if match:
      core_number_str = match.group()
      core_number = int(core_number_str)
    else:
      core_number = 2
    if core_number == 3:
      return 3
    elif core_number in (5, 7, 9):
      return core_number
    else:
      return 2
  elif processor_name == "Apple M2":
    return 9
  else:
    return 2

with open(input_file, 'r', encoding='utf-8') as csvfile, open(output_file, 'w', encoding='utf-8') as outfile:
    reader = csv.DictReader(csvfile)
    writer = csv.DictWriter(outfile, fieldnames=reader.fieldnames)
    writer.writeheader()
    for row in reader:
      cpu = row['cpu']
      score = processor_score(cpu)
      row['cpu'] = score
      writer.writerow(row)
```

Labeling GPU dengan range 1-9

```python
input_file = "/content/cleaned_scored_cpu.csv"
output_file = "cleaned_score_cpu-gpu.csv"

import csv
import re

input_file = "cleaned_scored_cpu.csv"
output_file = "cleaned_score_cpu-gpu.csv"

def clean_gpu_name(gpu_name):
    gpu_name = re.sub(r'\s*\(.*?\)', '', gpu_name).strip()
    return gpu_name

def gpu_score(gpu_name):
    gpu_name = clean_gpu_name(gpu_name).replace('"', '')

    if 'Mali-G72' in gpu_name:
        return 1
    if 'Imagination PowerVR GX6250' in gpu_name:
        return 2
    if re.search(r'(AMD Radeon (610M|660M|680M|760M|780M|R2|R4|R5|R7|RX Vega)|Intel (HD Graphics|UHD Graphics|Iris)|NVIDIA GeForce (MX|GTX 860M))', gpu_name):
        return 3
    if re.search(r'(AMD Radeon RX (6500M|6650M)|NVIDIA GeForce (GTX 1050|GTX 1650|RTX 2050)|Quadro (P520|T500|T550)|T600)', gpu_name):
        return 5
    if re.search(r'(Intel Arc A370M|AMD Radeon RX 6700M|NVIDIA (T1200|GeForce (GTX 1660|RTX (2060|2070|3500|3050|4000|4050))|RTX A1000))', gpu_name):
        return 6
    if re.search(r'(AMD Radeon RX 6800M|NVIDIA GeForce (RTX 2080|RTX 3060|RTX 3070|RTX 4060)|NVIDIA RTX A2000|Apple M2 GPU)', gpu_name):
        return 7
    if re.search(r'(NVIDIA (GeForce RTX (3080|4070|4080))|NVIDIA RTX A3000)', gpu_name):
        return 8
    if 'NVIDIA GeForce RTX 4090' in gpu_name:
        return 9

    return 'n/a'

with open(input_file, 'r', encoding='utf-8') as csvfile, open(output_file, 'w', encoding='utf-8', newline='') as outfile:
    reader = csv.DictReader(csvfile)
    writer = csv.DictWriter(outfile, fieldnames=reader.fieldnames)
    writer.writeheader()
    for row in reader:
        gpu = row['gpu']
        score = gpu_score(gpu)
        row['gpu'] = score
        writer.writerow(row)
```

Labeling RAM dengan range 1-11

```python
input_file = "/content/cleaned_score_cpu-gpu.csv"
output_file = "cleaned_score_cpu-gpu-ram.csv"

def ram_score(ram_capacity):
    ram_capacity_numeric = int(ram_capacity.replace('GB', '').strip())

    if ram_capacity_numeric <= 4:
        return 1
    elif 4 < ram_capacity_numeric <= 8:
        return 2
    elif 8 < ram_capacity_numeric <= 12:
        return 3
    elif 12 < ram_capacity_numeric <= 16:
        return 4
    elif 16 < ram_capacity_numeric <= 20:
        return 5
    elif 20 < ram_capacity_numeric <= 24:
        return 6
    elif 24 < ram_capacity_numeric <= 32:
        return 7
    elif 32 < ram_capacity_numeric <= 36:
        return 8
    elif 36 < ram_capacity_numeric <= 40:
        return 9
    elif 40 < ram_capacity_numeric <= 48:
        return 10
    else:
        return 11

with open(input_file, 'r', encoding='utf-8') as csvfile, open(output_file, 'w', encoding='utf-8', newline='') as outfile:
    reader = csv.DictReader(csvfile)
    writer = csv.DictWriter(outfile, fieldnames=reader.fieldnames)
    writer.writeheader()
    for row in reader:
        ram = row['ram']
        score = ram_score(ram)
        row['ram'] = score
        writer.writerow(row)
```

Labeling Disc Capacity dengan range 1-7

```python
import re
import csv

def split_disc(disc_description):
    disc_description = disc_description.lower()

    hdd_capacity = None
    ssd_capacity = None

    if 'hdd' in disc_description:
        match = re.search(r'(\d+)gb hdd', disc_description)
        if match:
            hdd_capacity = int(match.group(1))

    if 'ssd' in disc_description:
        match = re.search(r'(\d+)gb ssd', disc_description)
        if match:
            ssd_capacity = int(match.group(1))

    return hdd_capacity, ssd_capacity

input_file = "/content/cleaned_score_cpu-gpu-ram.csv"
output_file = "cleaned_score_cpu-gpu-ram-disc.csv"

def capacity_score(capacity):
    if capacity is None:
        return 0
    if capacity <= 128:
        return 1
    elif capacity <= 256:
        return 2
    elif capacity <= 512:
        return 3
    elif capacity <= 1000:
        return 4
    elif capacity <= 2000:
        return 5
    elif capacity <= 4000:
        return 6
    else:
        return 7

with open(input_file, 'r', encoding='utf-8') as csvfile, open(output_file, 'w', encoding='utf-8', newline='') as outfile:
    reader = csv.DictReader(csvfile)
    fieldnames = reader.fieldnames + ['HDD', 'SSD']
    writer = csv.DictWriter(outfile, fieldnames=fieldnames)
    writer.writeheader()
    for row in reader:
        disc = row['disc']
        hdd, ssd = split_disc(disc)
        hdd_score = capacity_score(hdd)
        ssd_score = capacity_score(ssd)
        row['HDD'] = hdd_score
        row['SSD'] = ssd_score
        del row['disc']
        writer.writerow(row)
```

## Analisis Data

Jumlah sebaran laptop

Load data dari CSV

```python
data = pd.read_csv('clean_laptop_data.csv')
```

Tambah Filter Harga, Edit bobot untuk kriteria dan produsen Laptop

```python
# Bobot Kriteria
criteria = ['producer', 'cpu', 'gpu', 'ram', 'HDD', 'SSD']
criteria_weights = {'producer' : 3,
                    'cpu'      : 7,
                    'gpu'      : 7,
                    'ram'      : 4,
                    'HDD'      : 0,
                    'SSD'      : 3
                    }

# Bobot Produser Laptop
producer_weights = {'HP'        : 2,
                    'Lenovo'    : 6,
                    'Acer'      : 4,
                    'Dell'      : 2,
                    'ASUS'      : 3,
                    'MSI'       : 1,
                    'Apple'     : 5,
                    'Microsoft' : 1
                    }
```

Normalisasi Data

```python
# Normalisasi data produser laptop
total_producer_weight = sum(producer_weights.values())
normalized_producer_weights = {producer: weight / total_producer_weight for producer, weight in producer_weights.items()}

# Normalisasi bobot
total_criteria_weight = sum(criteria_weights.values())
normalized_criteria_weights = {criterion: weight / total_criteria_weight for criterion, weight in criteria_weights.items()}

# Normalisasi data kriteria
for criterion in criteria[1:]:
    data[criterion] = data[criterion] / data[criterion].max()
```

Kalkulasi skor menggunakan Metode AHP

```python
for criterion in criteria:
    if criterion == 'producer':
        data[criterion + '_score'] = data[criterion].map(normalized_producer_weights)
    else:
        data[criterion + '_score'] = data[criterion]

data['ahp_score'] = data[[criterion + '_score' for criterion in criteria]].prod(axis=1)
```

Kalkulasi skor menggunakan Metode WP

```python
weights = [normalized_criteria_weights[criterion] for criterion in criteria]
benefits = [criterion + '_score' for criterion in criteria]
cost = 'price_in_dollar'

data['wp_score'] = data[benefits].apply(lambda x: np.prod(x ** weights), axis=1) / data[cost]
```

Mengurutkan data/ranking dari skor Metode AHP dan WP

```python
data['ahp_rank'] = data['ahp_score'].rank(ascending=False)
data['wp_rank'] = data['wp_score'].rank(ascending=False)
```

Final Decision untuk kedua metode menggunkanan metode rata-rata

```python
data['combined_rank'] = (data['ahp_rank'] + data['wp_rank']) / 2
data.sort_values(by='combined_rank', inplace=True)
```
